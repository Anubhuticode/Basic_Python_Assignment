```python
from datetime import date 
today=datetime.date.today()

file=open((today.strftime("%d %B %y")+".txt"),'w')
with open(today.strftime("%d %B %y"),'w') as file:
    file.write(str(today))
    file.close()
with open(today.strftime("%d %B %y"),"r") as file:
    print(file.readline())
```
